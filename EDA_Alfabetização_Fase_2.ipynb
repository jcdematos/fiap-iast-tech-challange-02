```python
uf = pd.read_csv('br_inep_avaliacao_alfabetizacao_uf.csv')
uf.head()
```

```python
uf.shape
```

```python
uf.info()
```

```python
# Definir as colunas chave
colunas_chave = ["ano", "sigla_uf"]

# 1. Verificar se existem duplicados considerando apenas as chaves
tem_duplicados = uf.duplicated(subset=colunas_chave).any()
print(f"O CSV possui dados duplicados para a combinação de Ano e UF? {'Sim' if tem_duplicados else 'Não'}")

# 2. Contar a quantidade de linhas duplicadas com base nessas chaves
total_duplicados = uf.duplicated(subset=colunas_chave).sum()
print(f"Quantidade de registros duplicados para Ano e UF: {total_duplicados}")
```

```python
# Conta os valores vazios em cada coluna do DataFrame uf
vazios_por_coluna_uf = uf.isna().sum()
print("Células vazias por coluna no DataFrame 'uf':")
print(vazios_por_coluna_uf)
```
